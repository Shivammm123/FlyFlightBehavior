```python



# Convert lists to NumPy arrays if they aren't already arrays
AngV_J74_Gal4 = np.array(AngV_J74_Gal4)
AngV_J73_Gal4 = np.array(AngV_J73_Gal4)
AngV_J64 = np.array(AngV_J64)
AngV_J72_Gal4 = np.array(AngV_J72_Gal4)

# Flatten the arrays if needed (if they are multidimensional)
AngV_J74_Gal4 = AngV_J74_Gal4.flatten()
AngV_J73_Gal4 = AngV_J73_Gal4.flatten()
AngV_J64 = AngV_J64.flatten()
AngV_J72_Gal4 = AngV_J72_Gal4.flatten()

# Perform pairwise t-tests
t_stat_73_74, p_value_73_74 = stats.ttest_ind(AngV_J73_Gal4, AngV_J74_Gal4)
t_stat_74_72, p_value_74_72 = stats.ttest_ind(AngV_J74_Gal4, AngV_J72_Gal4)
t_stat_72_64, p_value_72_64 = stats.ttest_ind(AngV_J72_Gal4, AngV_J64)
t_stat_73_64, p_value_73_64 = stats.ttest_ind(AngV_J73_Gal4, AngV_J64)

# Print t-statistic and p-values
print(f"T-test between J73 and J74: t-stat = {t_stat_73_74:.2f}, p-value = {p_value_73_74:.2e}")
print(f"T-test between J74 and J72: t-stat = {t_stat_74_72:.2f}, p-value = {p_value_74_72:.2e}")
print(f"T-test between J72 and J64: t-stat = {t_stat_72_64:.2f}, p-value = {p_value_72_64:.2e}")
print(f"T-test between J73 and J64: t-stat = {t_stat_73_64:.2f}, p-value = {p_value_73_64:.2e}")


# Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(AngV_J74_Gal4, AngV_J73_Gal4, AngV_J64, AngV_J72_Gal4)

# Print F-statistic and p-value
print(f"ANOVA: F-statistic = {f_stat:.2f}, p-value = {p_value:.2e}")
```

```python
# Number of comparisons
n_tests = 4

# Original significance level
alpha = 0.05

# Bonferroni corrected significance level
alpha_corrected = alpha / n_tests

print(f"Bonferroni-corrected significance level: {alpha_corrected:.4f}")

# Checking if the p-values are significant after Bonferroni correction
p_values = [p_value_73_74, p_value_74_72, p_value_72_64, p_value_73_64]

# Results after Bonferroni correction
for i, p_value in enumerate(p_values):
    is_significant = p_value < alpha_corrected
    print(f"T-test {i+1} p-value = {p_value:.2e}, Significant after Bonferroni correction: {is_significant}")
```
